Scenario

You are building a screening tool for a clinical trial.

The system should determine whether a patient is eligible to participate based on several criteria.

Collect the following information:
Patient age
Gender (M/F)
Weight (kg)
Has Diabetes? (Y/N)
Has Heart Disease? (Y/N)
Pregnant? (Y/N)
Eligibility Rules
Rule 1: Age

Patient must be between 18 and 65 years inclusive.

If not, immediately display:

Not Eligible: Age criteria not met.
Rule 2: Weight

Patient must weigh at least 50 kg.

If not:

Not Eligible: Weight criteria not met.
Rule 3: Medical History

Patient cannot have both:

Diabetes = Y
Heart Disease = Y

If both are present:

Not Eligible: High-risk medical history.
Rule 4: Pregnancy

If gender is F and pregnant is Y:

Not Eligible: Pregnancy exclusion criterion met.
Final Decision

If all checks pass:

Eligible for Study Enrollment.
Bonus Challenge

Calculate a risk score:

Condition	Points
Age > 55	+2
Diabetes	+3
Heart Disease	+4
Weight < 60 kg	+2

Then classify:

0–2 → Low Risk
3–5 → Moderate Risk
6+ → High Risk

Display:

Risk Score: 5
Risk Category: Moderate Risk

```python

Patient_age = int(input("What is your age"))

if Patient_age <18 or Patient_age >65:
    print("Not Eligible: Age criteria not met.")
else:
    Weight_kg = int(input ("What is your weight"))

    if Weight_kg <50:
        print("Not Eligible: Weight criteria not met.")
    else:
        Has_Diabetes = input ("Do you suffer from Diabetes (Y/N) ")
        Has_Heart_Disease = input ("Do you have heart Disease? (Y/N)")

        if Has_Diabetes == "Y" and Has_Heart_Disease == "Y":
            print("Not Eligible: High-risk medical history.")
        else:
            Gender  = input ("Male or Female (M//F)")
            Pregnant = input("Are you pregnant (Y/N)")
            
            if Gender == "F" and Pregnant == "Y":
                print("Not Eligible: Pregnancy exclusion criterion met.")
            else:
               print("Eligible for Study Enrollment.")
```

```python
#Calculate risk score
points = 0
if Patient_age > 55:
    points =+2
if Has_Diabetes == "Y":
    points =+3

if Has_Heart_Disease == "Y":
    points =+ 4

if Weight_kg <60:
    points =+2

print(f"Risk Score: {points}")
if points <=2:
    print("Risk Category: Low risk")
elif points <=5:
    print("Risk Category: Moderate risk")
else:
    print("Risk Category High risk")
    
```
